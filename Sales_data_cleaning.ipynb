```python
import pandas as pd

# Load the uploaded CSV file
df = pd.read_csv('customer_orders.csv')

# Preview the data
print("Initial Data Snapshot:")
print(df.head())

# Drop rows with missing values
df_cleaned = df.dropna()

# Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# Rename columns for clarity
df_cleaned = df_cleaned.rename(columns={
    'Client_ID': 'Customer_ID',
    'Item': 'Product',
    'Units': 'Quantity',
    'Amount': 'Revenue'
})

# Confirm column names
print("\nRenamed Columns:")
print(df_cleaned.columns)

# Convert Revenue to numeric
df_cleaned['Revenue'] = df_cleaned['Revenue'].replace('[\\$,]', '', regex=True).astype(float)

# Filter rows: only include sales over $100
df_filtered = df_cleaned[df_cleaned['Revenue'] > 100]

# Export cleaned data to new CSV
df_filtered.to_csv('cleaned_customer_orders.csv', index=False)
print("\nCleaned data exported successfully.")
```
